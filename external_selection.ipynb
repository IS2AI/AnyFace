```python
# import the necessary packages

import matplotlib.patches as patches
import matplotlib.pyplot as plt
from imutils import paths
from PIL import Image
import numpy as np
import imutils
import random
import shutil
import csv
import cv2
import os
```

```python
def make_dir(dirName):
    # Create a target directory & all intermediate 
    # directories if they don't exists
    
    if not os.path.exists(dirName):
        os.makedirs(dirName, exist_ok = True)
        print("[INFO] Directory " ,dirName,  " created")
    else:
        print("[INFO] Directory " ,dirName,  " already exists") 
```

```python
ROOT = 'selected-external'    # directory to save selected images for the external datasets
```

```python
dataset = 'animefaces'                # dataset name ['tj', 'animefaces', 'artisticfaces', 'metfaces', 'turtles']

# Make new directories for images and labels
make_dir(os.path.join(ROOT, dataset, 'images'))
make_dir(os.path.join(ROOT, dataset, 'labels_eval'))

# Get the image and label paths
imagesList = sorted(list(paths.list_images(os.path.join('dataset', dataset, 'images'))))
labelsList = sorted(list(paths.list_files(os.path.join('dataset', dataset, 'labels_eval'), validExts='txt')))

remove = []   # list of images to remove
N = 0

# Iterate over label files
for labelPath in labelsList:
    labelFile = open(labelPath, "r")
    
    n = len(labelFile.readlines())    # the number of faces
    print('Number of faces: {}'.format(n))
    
    # Add the path image without faces to remove list
    if n == 0:
        remove.append(labelPath)
        N += 1
print(N)

# Remove image and labels without faces
for r in remove:
    labelsList.remove(r)
    imagesList.remove(r.replace('txt', 'jpg').replace('labels_eval', 'images'))
    print('Removed: ', r)

# Check if number of images and labels match
if len(imagesList) != len(labelsList):
    print(len(imagesList), len(labelsList))
    raise Exception("Number of images and labels does not match")

# Generate a list random 160 integers from 0 to dataset size
randomIdx = random.sample(range(0, len(imagesList)), 160)

# Copy 160 random images and labels to selected directory
for idx in randomIdx:
    desImage = os.path.join(ROOT, dataset, 'images', imagesList[idx].split('/')[-1])
    desLabel = os.path.join(ROOT, dataset, 'labels_eval', imagesList[idx].split('/')[-1].split('.')[0] + '.txt')

    print(desImage, desLabel)

    shutil.copyfile(imagesList[idx], desImage)
    shutil.copyfile(labelsList[idx], desLabel)
```

```python
dataset = 'pets'    # dataset name

# Make new directories for images and labels
make_dir(os.path.join(ROOT, dataset, 'images'))
make_dir(os.path.join(ROOT, dataset, 'labels_eval'))

# Get the image paths
imagesListAll = list(paths.list_images(os.path.join('dataset', dataset, 'images')))

# For each species available, select 5 images and corresponding labels
imagesList = []
labelsList = []
d = {}

for imagePath in imagesListAll:
    filename = imagePath.split('/')[-1].split('.')[0]
    species = '_'.join(filename.split('_')[:-1])
    
    if species not in d.keys():    # if species was not encountered, add to dictionary
        d[species] = 1
    else:
        d[species] += 1            # if species was encountered, update count in the dictionary
        
    if d[species] < 6:             # if the number of images corresponding to species is less than 5, add
        imagesList.append(imagePath)
        labelsList.append(imagePath.replace('images', 'labels_eval').replace('jpg', 'txt'))
        
        
# Check if number of images and labels match
if len(imagesList) != len(labelsList):
    raise Exception("Number of images and labels does not match")

# Generate a list random 160 integers from 0 to dataset size
randomIdx = random.sample(range(0, len(imagesList)), 160)

# Copy 160 random images and labels to selected directory
for idx in randomIdx:
    desImage = os.path.join(ROOT, dataset, 'images', imagesList[idx].split('/')[-1])
    desLabel = os.path.join(ROOT, dataset, 'labels_eval', imagesList[idx].split('/')[-1].split('.')[0] + '.txt')

    print(desImage, desLabel)

    shutil.copyfile(imagesList[idx], desImage)
    shutil.copyfile(labelsList[idx], desLabel)
```
