```python
# import the necessary packages
from shutil import copyfile
from imutils import paths
import os
```

```python
def make_dir(dirName):
    # Create a target directory & all intermediate 
    # directories if they don't exists
    if not os.path.exists(dirName):
        os.makedirs(dirName, exist_ok = True)
        print("[INFO] Directory " , dirName ,  " created")
    else:
        print("[INFO] Directory " , dirName ,  " already exists")
```

```python
SET = 'test'

SOURCE_PATH = os.path.join('dataset/tfw/', SET)
TARGET_PATH = os.path.join('dataset/tfw_split/', SET)

make_dir(TARGET_PATH)

make_dir(os.path.join(TARGET_PATH, 'indoor', 'images'))
make_dir(os.path.join(TARGET_PATH, 'indoor', 'labels'))

make_dir(os.path.join(TARGET_PATH, 'outdoor', 'images'))
make_dir(os.path.join(TARGET_PATH, 'outdoor', 'labels'))
```

```python
# get path to the images and annotations
annotPaths = list(paths.list_files(os.path.join(SOURCE_PATH, 'labels_eval'),  validExts="txt"))
annotPaths = sorted(annotPaths)

imagePaths = list(paths.list_files(os.path.join(SOURCE_PATH, 'images'),  validExts="png"))
imagePaths = sorted(imagePaths)

print(len(annotPaths), len(imagePaths))
```

```python
count = 1

for imagePath, annotPath in zip(imagePaths, annotPaths):
    imageName = imagePath.split('/')[-1]
    
    if len(imageName) <= 8:
        env = 'outdoor'
    else:
        env = 'indoor'

    copyfile(imagePath, os.path.join(TARGET_PATH, env, 'images', imagePath.split('/')[-1]))
    copyfile(annotPath, os.path.join(TARGET_PATH, env, 'labels', annotPath.split('/')[-1]))
    
    print("[INFO] Copying files {}/{}".format(count, len(imagePaths)))
    count +=1
```
